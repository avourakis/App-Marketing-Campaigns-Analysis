# App Marketing Campaigns Analysis (Visualizations)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
sns.set()
sns.set_style("whitegrid")
```

### Q1 A

```python
q1_a = pd.read_csv("Data/q1_a.csv")
q1_a.plot(x="date", y="count", legend = False, figsize=(20,12), color="b")
plt.title("Number of installs per Day")
plt.xlabel("Dates")
plt.ylabel("Number of Installs")
#plt.savefig("Plots/q1_a.png", dpi=150)
```

### Q1 B

```python
q1_b = pd.read_csv("Data/q1_b.csv")
q1_b.sort_values(by="num_installs", ascending=True, inplace=True)
q1_b.plot.bar(x="campaign", y="num_installs", legend=False, figsize=(20,12), color="b")
plt.title("Number of installs per Campaign")
plt.xlabel("Campaigns")
plt.ylabel("Number of Installs")
#plt.savefig("Plots/q1_b.png", dpi=150)
```

```python
q1_b["num_installs"]/np.sum(q1_b["num_installs"])*100
```

### Q1 C

```python
q1_c = pd.read_csv("Data/q1_c.csv")
q1_c.sort_values(by="num_installs", ascending=True, inplace=True)
q1_c.plot.bar(x="network", y="num_installs", legend=False, figsize=(20,12), color="b", rot=0)
plt.title("Number of installs per Network")
plt.xlabel("Networks")
plt.ylabel("Number of Installs")
#plt.savefig("Plots/q1_c.png", dpi=150)
```

### Q4 A

### Q4 B

### Q5 A

```python
q5_a = pd.read_csv("Data/q5_a.csv")
q5_a.sort_values(by="roi", ascending=True, inplace=True)
positive_values = q5_a["roi"] > 0
q5_a.plot.bar(x="campaign", y="roi", legend=False, figsize=(20,12), color=positive_values.map({True: 'b', False: 'r'}))
plt.title("Intial ROI per Campaign")
plt.xlabel("Campaigns")
plt.ylabel("ROI")
#plt.savefig("Plots/q5_a.png", dpi=150)
```

### Q5 B

```python
q5_b = pd.read_csv("Data/q5_b.csv")
q5_b.sort_values(by="roi", ascending=True, inplace=True)
positive_values = q5_b["roi"] > 0
q5_b.plot.bar(x="network", y="roi", legend=False, figsize=(20,12), rot=0, color=positive_values.map({True: 'b', False: 'r'}))
plt.title("Initial ROI per Network")
plt.xlabel("Networks")
plt.ylabel("ROI")
#plt.savefig("Plots/q5_b.png", dpi=150)
```
